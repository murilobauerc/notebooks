```python
import pandas as pd
%matplotlib inline

df_census = pd.read_csv('census_income_data.csv')
df_census.info()
```

```python
df_census.hist(figsize=(8,8));
```

```python
df_census['education-num'].plot(kind='hist');
# usando a função mais geral para gráficos 'plot'
```

```python
df_census['education-num'].value_counts()
# value.counts() - contagem para cada valor/barra do gráfico
```

```python
 df_census['education-num'].value_counts().plot(kind = 'bar');
# visualização dessa contagem em um gráfico de barras
```

```python
# criando um gráfico de pizza para a coluna education-num
df_census['education-num'].value_counts().plot(kind = 'pie', figsize = (6,6));
```
