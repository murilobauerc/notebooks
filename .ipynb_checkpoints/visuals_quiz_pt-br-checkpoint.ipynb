```python
# importa e carrega dados
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df_powerplant = pd.read_csv('powerplant_data.csv')
df_powerplant.info()

# AT = Ambient Temperature
# V = Exhaust Vacuum
# AP = Ambient Pressure
# RH = Relative Humidity
# PE = Energy Output
```

```python
# trace um gráfico com a relação entre temperatura e saída elétrica
# df_powerplant['AT'].plot(kind = 'hist');
df_powerplant['AT'].plot(kind='hist'); # blue plot
df_powerplant['PE'].plot(kind='hist'); # orange plot
```

```python
# trace um gráfico com a distribuição da umidade
df_powerplant['RH'].plot(kind='hist');
```

```python
# trace um gráfico de caixas para cada variável
df_powerplant.boxplot();
```
